```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the dataset
df = pd.read_csv('C:/Users/viina/Downloads/car.csv')

df
```

```python
# a) Calculate the average speed and distance
average_speed = df['speed'].mean()
average_distance = df['dist'].mean()

# b) Calculate the median and midrange
median_speed = df['speed'].median()
median_distance = df['dist'].median()

midrange_speed = (df['speed'].min() + df['speed'].max()) / 2
midrange_distance = (df['dist'].min() + df['dist'].max()) / 2

# c) Find the mode and comment on the data modality
mode_speed = df['speed'].mode()
mode_distance = df['dist'].mode()

speed_modality = "Unimodal" if len(mode_speed) == 1 else "Bimodal or Multimodal"
distance_modality = "Unimodal" if len(mode_distance) == 1 else "Bimodal or Multimodal"

# d) Calculate the variance and standard deviation
variance_speed = df['speed'].var()
variance_distance = df['dist'].var()

std_deviation_speed = df['speed'].std()
std_deviation_distance = df['dist'].std()

# e) Find the five-number summary
five_num_summary_speed = df['speed'].describe(percentiles=[.25, .5, .75])
five_num_summary_distance = df['dist'].describe(percentiles=[.25, .5, .75])

# f) Show the histogram and boxplot
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['speed'], kde=True)
plt.title('Histogram of Speed')
plt.xlabel('Speed')

plt.subplot(1, 2, 2)
sns.histplot(df['dist'], kde=True)
plt.title('Histogram of Distance')
plt.xlabel('Distance')

plt.show()

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['speed'])
plt.title('Boxplot of Speed')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['dist'])
plt.title('Boxplot of Distance')

plt.show()

# Print results
print("a) What are the average speed and the distance of the cars?")
print(f"   Average Speed: {average_speed}")
print(f"   Average Distance: {average_distance}\n")

print("b) What is the median and midrange of the data?")
print(f"   Median Speed: {median_speed}")
print(f"   Median Distance: {median_distance}")
print(f"   Midrange Speed: {midrange_speed}")
print(f"   Midrange Distance: {midrange_distance}\n")

print("c) Find the mode of the data and comment on the data modality (i.e., unimodal or bimodal).")
print(f"   Mode Speed: {mode_speed.values}")
print(f"   Speed Modality: {speed_modality}")
print(f"   Mode Distance: {mode_distance.values}")
print(f"   Distance Modality: {distance_modality}\n")

print("d) What are the variance and the standard deviation of the data?")
print(f"   Variance of Speed: {variance_speed}")
print(f"   Variance of Distance: {variance_distance}")
print(f"   Standard Deviation of Speed: {std_deviation_speed}")
print(f"   Standard Deviation of Distance: {std_deviation_distance}\n")

print("e) Find the five-number summary of the data.")
print(f"   Five Number Summary of Speed:")
print(f"     Minimum: {five_num_summary_speed['min']}")
print(f"     25th Percentile: {five_num_summary_speed['25%']}")
print(f"     Median: {five_num_summary_speed['50%']}")
print(f"     75th Percentile: {five_num_summary_speed['75%']}")
print(f"     Maximum: {five_num_summary_speed['max']}")
print(f"   Five Number Summary of Distance:")
print(f"     Minimum: {five_num_summary_distance['min']}")
print(f"     25th Percentile: {five_num_summary_distance['25%']}")
print(f"     Median: {five_num_summary_distance['50%']}")
print(f"     75th Percentile: {five_num_summary_distance['75%']}")
print(f"     Maximum: {five_num_summary_distance['max']}\n")

print("f) Show the histogram and boxplot of the data.")
```
